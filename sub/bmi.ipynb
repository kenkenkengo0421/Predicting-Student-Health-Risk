## BMIについて

* BMI（Body Mass Index）は、身長と体重から算出される体格指数で、肥満や低体重の判定に用いられる国際的な指標。

###  **BMI** = 体重（kg） ÷ 身長（m）²
身長160cm、体重53kgの場合、53 ÷ (1.6 × 1.6) = 20.7

肥満度の判定基準

* 18.5未満：低体重（痩せ）
* 18.5〜24.9：標準体重
* 25以上：肥満
* 世界保健機関（WHO）ではBMI30以上を肥満としていますが、日本人は欧米人より平均BMIが低いため、25以上を肥満とする基準が一般的

BMIは、生活習慣病のリスク評価やダイエット・体型管理の目安として活用できます。標準体重はBMI22を基準に算出され、病気になりにくい体重の目安とされています。BMIだけでなく、体脂肪率や筋肉量も併せて確認することで、より正確な健康管理が可能です


```python
matplotlib.style.use('bmh')
print("全体の心拍数")
plt.figure(figsize=(10, 10))
plt.hist(df['bmi'].dropna(), bins=30,edgecolor='white')
plt.title("bmi")
plt.xticks(np.arange(int(df['bmi'].min()), int(df['bmi'].max()) + 1, 1))
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.histplot(data=df, x='bmi', hue='health_condition', multiple='stack')
plt.xticks(np.arange(int(df['bmi'].min()), int(df['bmi'].max()) + 1, 1))
plt.show()
```

* fit（紫）: 健康的な状態
* at-risk（赤）: リスクあり（予備軍）
* unhealthy（水色）: 不健康な状態

```python
plt.figure(figsize=(15, 5))
sns.kdeplot(data=df, x='bmi', hue='health_condition', fill=True, common_norm=False)
plt.title('bmi')
plt.xticks(np.arange(int(df['bmi'].min()), int(df['bmi'].max()) + 1, 1))
plt.show()
```

* 27以降が不健康な傾向にある
* 19以下は健康、不健康な人がほぼいない。
