```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
```

```python
df
```

```python
df['heart_rate']
```

```python
matplotlib.style.use('bmh')
print("全体の心拍数")
plt.figure(figsize=(10, 10))
plt.hist(df['heart_rate'].dropna(), bins=30,edgecolor='white')
plt.title("heart_rate")
plt.show()
```

* 心拍数が多いピークゾーンは70から80くらい。
それくらいを基準にする

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='heart_rate', hue='health_condition', multiple='stack')
plt.xticks()
plt.show()
```

* fit（紫）: 健康的な状態
* at-risk（赤）: リスクあり（予備軍）
* unhealthy（水色）: 不健康な状態

```python
plt.figure(figsize=(10, 5))
sns.kdeplot(data=df, x='heart_rate', hue='health_condition', fill=True, common_norm=False)
plt.title('')
plt.show()
```

* 心拍数はあまり関連性がなさそう
