# 心拍数

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style
import seaborn as sns
from my_plot import Hist, sns_Hist, sns_kde
```

```python
df = pd.read_csv("../data/train.csv")
```

```python
df
```

```python
df['heart_rate']
```

```python
matplotlib.style.use('bmh')
print("全体の心拍数")
Hist(data=df,column_name='heart_rate',figsize=(10, 10),step=10)
```

* 心拍数が多いピークゾーンは70から80くらい。
それくらいを基準にする

```python
sns_Hist(data=df,x='heart_rate',hue='health_condition',figsize=(10, 5),step=10)
```

* fit（紫）: 健康的な状態
* at-risk（赤）: リスクあり（予備軍）
* unhealthy（水色）: 不健康な状態

```python
sns_kde(data=df,x='heart_rate',hue='health_condition',figsize=(10, 5),step=10)
```

* 心拍数はあまり関連性がなさそう
