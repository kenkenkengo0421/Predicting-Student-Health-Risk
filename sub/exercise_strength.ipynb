# 運動強度

$$\text{運動強度} = \frac{\text{消費カロリー}}{(\text{運動時間} + 0.1) \times \text{BMI}}$$



```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style
import seaborn as sns
from my_plot import Hist, sns_Hist, sns_kde
```

```python
df = pd.read_csv("train.csv")
```

```python
df
```

```python
df["exercise_strength"] = df['calorie_expenditure'] / ((df['exercise_duration'] + 0.1) * df['bmi'])
```

```python
df["exercise_strength"]
```

```python
# 例：値が10未満のデータだけにする
df_filtered = df[df['exercise_strength'] < 10]
sns_kde(data=df_filtered, x='exercise_strength',hue='health_condition', figsize=(10, 10), step=0.5)
```
