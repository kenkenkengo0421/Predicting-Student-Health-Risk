### カロリー消費量について
* 1日の平均消費カロリーは、性別や年齢、活動量によって異なりますが、

    一般的な成人で約1,800〜2,500kcalです

    内訳は、じっとしていても消費される基礎代謝量（女性で約1,100〜1,300kcal、男性で約1,400〜1,

    600kcal）に、生活や運動の活動量が加わって決まります。

```python
matplotlib.style.use('bmh')
print("カロリー消費量")
Hist(data=df,column_name='calorie_expenditure',figsize=(10, 5),step=200)
```

```python
sns_Hist(data=df,x='calorie_expenditure',hue='health_condition',figsize=(10, 5),step=200)
```

* fit（紫）: 健康的な状態
* at-risk（赤）: リスクあり（予備軍）
* unhealthy（水色）: 不健康な状態

```python
sns_kde(data=df,x='calorie_expenditure',hue='health_condition',figsize=(10, 5),step=200)
```

* 1400以下は不健康の傾向あり
* 1400kcalは一般的に、運動を含めた1日の摂取カロリーの目安
それ以下の方たちが不健康の傾向あり
