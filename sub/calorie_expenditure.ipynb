### カロリー消費量について
* 1日の平均消費カロリーは、性別や年齢、活動量によって異なりますが、

    一般的な成人で約1,800〜2,500kcalです

    内訳は、じっとしていても消費される基礎代謝量（女性で約1,100〜1,300kcal、男性で約1,400〜1,

    600kcal）に、生活や運動の活動量が加わって決まります。

```python
matplotlib.style.use('bmh')
print("カロリー消費量")
plt.figure(figsize=(10, 5))
plt.hist(df['calorie_expenditure'].dropna(), bins=30,edgecolor='white')
plt.title("calorie_expenditure")
plt.xticks(np.arange(int(df['calorie_expenditure'].min()), int(df['calorie_expenditure'].max()) + 1, 200))
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='calorie_expenditure', hue='health_condition', multiple='stack')
plt.xticks(np.arange(int(df['calorie_expenditure'].min()), int(df['calorie_expenditure'].max()) + 1, 200))
plt.show()
```

* fit（紫）: 健康的な状態
* at-risk（赤）: リスクあり（予備軍）
* unhealthy（水色）: 不健康な状態

```python
plt.figure(figsize=(10, 5))
sns.kdeplot(data=df, x='calorie_expenditure', hue='health_condition', fill=True, common_norm=False)
plt.title('')
plt.xticks(np.arange(int(df['calorie_expenditure'].min()), int(df['calorie_expenditure'].max()) + 1, 200))
plt.show()
```

* 1400以下は不健康の傾向あり
* 1400kcalは一般的に、運動を含めた1日の摂取カロリーの目安
それ以下の方たちが不健康の傾向あり
