# ストレス係数:ストレスレベル/睡眠時間

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style
import seaborn as sns
import numpy as np
from my_plot import Hist, sns_Hist, sns_kde
```

```python
df = pd.read_csv("../data/train.csv")
```

```python
df
```

```python
# 実際のデータに合わせたマッピング
stress_mapping = {'low': 1, 'medium': 2, 'high': 3}
df['stress_level_num'] = df['stress_level'].map(stress_mapping)
# ストレス負係数:  ストレスレベル/睡眠時間
df['stress_per_sleep'] = df['stress_level_num'] / df['sleep_duration']
```
