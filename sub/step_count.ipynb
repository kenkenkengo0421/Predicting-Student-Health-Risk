```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("../data/train.csv")
```

```python
df
```

```python
df['step_count']
```

```python
matplotlib.style.use('bmh')
print("ステップ数")
plt.figure(figsize=(30, 5))
plt.hist(df['step_count'].dropna(), bins=30,edgecolor='white')
plt.title("step_count")
plt.show()
```

```python
plt.figure(figsize=(30, 5))
sns.histplot(data=df, x='step_count', hue='health_condition', multiple='stack')
plt.show()
```

* fit（紫）: 健康的な状態
* at-risk（赤）: リスクあり（予備軍）
* unhealthy（水色）: 不健康な状態

5500歩より下回りだしたくらいから健康な人が減る。

#### 健康維持や長寿命化のための1日の理想的な歩数:

* 成人で**8,000歩**（うち早歩き20分）、健康寿命の延伸には**9,000歩**が目安です。必ずしも「1日1万歩」を達成しなくても、まずは「今より＋1,000歩（約10分）」増やすことから始めるのが効果的とされています。

```python
plt.figure(figsize=(30, 5))
sns.kdeplot(data=df, x='step_count', hue='health_condition', fill=True, common_norm=False)
plt.title('')
plt.show()
```
