```python
sns_Hist(data=df,x='step_count',hue='health_condition',figsize=((30, 5)),step=1000)
```

* fit（紫）: 健康的な状態
* at-risk（赤）: リスクあり（予備軍）
* unhealthy（水色）: 不健康な状態

5500歩より下回りだしたくらいから健康な人が減る。

#### 健康維持や長寿命化のための1日の理想的な歩数:

* 成人で**8,000歩**（うち早歩き20分）、健康寿命の延伸には**9,000歩**が目安です。必ずしも「1日1万歩」を達成しなくても、まずは「今より＋1,000歩（約10分）」増やすことから始めるのが効果的とされています。
